```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import FeatureUnion
from sklearn.model_selection import StratifiedKFold
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
DANE=r'Dane\ames2000_NAfix.csv'
dane=pd.read_csv(DANE)
dane.head(15)
```

```python
column_int = []
column_obj = []
i=0
for typ in dane.dtypes:
    if dane.dtypes[i]=='object':
        column_int.append(dane)
    
    elif dane.dtypes[i]=='int64':
        column_obj.append(dane)
    i+=1
```

```python
dane_copy= dane.copy()
dane_copy.head(2)
```

```python
X = dane_copy.drop(["SalePrice"], axis=1)
y = dane_copy["SalePrice"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Inspired from stackoverflow.com/questions/25239958
class MostFrequentImputer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.most_frequent_ = pd.Series([X[c].value_counts().index[0] for c in X],
                                        index=X.columns)
        return self
    def transform(self, X, y=None):
        return X.fillna(self.most_frequent_)
```

```python
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names]
```

```python
cat_pipeline = Pipeline([
        ("select_cat", DataFrameSelector(column_obj)),
        ("imputer", MostFrequentImputer()),
        ("cat_encoder", OneHotEncoder(sparse=False)),
        ('standardscaler', StandardScaler())
    ])
```

```python
cat_pipeline.fit_transform(X_train)
```

```python
num_pipeline = Pipeline([
        ("select_numeric", DataFrameSelector(column_int)),
        ("imputer", SimpleImputer(strategy="median")),
        ('standardscaler', StandardScaler())
    ])
```

```python
num_pipeline.fit_transform(X_train)
```

```python
preprocess_pipeline = FeatureUnion(transformer_list=[
        ("num_pipeline", num_pipeline),
        ("cat_pipeline", cat_pipeline),
    ])
```

```python
seed=123
kfold = StratifiedKFold(n_splits=5, random_state=seed)
```

```python
from sklearn.linear_model import LogisticRegression

pipe_1 = Pipeline([('preprocessing', preprocess_pipeline), ('classifier', LogisticRegression(C=1))])

param_grid_1 = {
            'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100]
}

grid_1 = GridSearchCV(pipe_1, param_grid_1, cv=kfold, return_train_score=True)

grid_1.fit(X_train, y_train)
#grid_1.best_params_


```

```python
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression(max_iter=1000)
lm.fit(X_train, y_train)
print("Test set accuracy: {:.2f}".format(lm.score(X_test, y_test)))
```
