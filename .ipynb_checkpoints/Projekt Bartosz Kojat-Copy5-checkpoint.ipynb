```python
dane_copy =dane.copy() # tworzę kopię na której bedę pracować
dane_copy.head(5)
```

```python
X = dane_copy.drop(["SalePrice"], axis=1) #ustawiam części testowe i treningowe
y = dane_copy["SalePrice"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
def select_cat_type(X): # tworzę dwie funkcje do selekcji kolumn ( dane kategoryczne i numeryczne)
    return X.select_dtypes(include='object')

def select_num_type(X):
    return X.select_dtypes(include=np.number)
```

```python
# tworzę pipeline do kolumn kategorycznych 
cat_pipeline = Pipeline([
        ("select_cat", FunctionTransformer(select_cat_type)),
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
        ("cat_encoder", OneHotEncoder(sparse=False,handle_unknown='ignore')),
        ('standardscaler', StandardScaler())
    ])
```

```python
# tworzę pipeline do kolumn numerycznych 
num_pipeline = Pipeline([
        ("select_numeric", FunctionTransformer(select_num_type)),
        ("imputer", SimpleImputer(strategy="mean")),
        ('standardscaler', StandardScaler())
    ])
```

```python
# łącze oba pipeline w jeden
preprocess_pipeline = FeatureUnion(transformer_list=[
        ("num_pipeline", num_pipeline),
        ("cat_pipeline", cat_pipeline),
    ])
```

```python
seed=123
kfold = StratifiedKFold(n_splits=5, random_state=seed)
```

```python
grid_0 = GridSearchCV(make_pipeline(preprocess_pipeline, PolynomialFeatures(degree=2), LinearRegression(n_jobs=-1)),
                    param_grid={'polynomialfeatures__degree': [1, 2]},
                    cv=kfold,
                    refit=True)
grid_0.fit(X, y)
print(grid_0.score(X_test, y_test))
print(grid_0.best_params_)
```

```python
grid_1 = GridSearchCV(make_pipeline(preprocess_pipeline, PolynomialFeatures(degree=2), Lasso()),
                    param_grid={'polynomialfeatures__degree': [1, 2,],
                    'lasso__alpha': [0.1, 1, 2, 3]},
                    cv=kfold,
                    refit=True)
grid_1.fit(X_train, y_train)
print(grid_1.score(X_test, y_test))
print(grid_1.best_params_)
```

```python
grid_2 = GridSearchCV(make_pipeline(preprocess_pipeline, PolynomialFeatures(degree=2),Ridge(alpha=1, tol=0.1)),
                    param_grid={'polynomialfeatures__degree': [1, 2,],
                    'ridge__alpha': [0.1, 1, 2, 4, 8, 20, 40]},
                    cv=5,
                    refit=True)



grid_2.fit(X_train, y_train)
grid_2.score(X_test, y_test)
print(grid_2.score(X_test, y_test))
print(grid_2.best_params_)
print(grid_2.best_estimator_)
```

```python
grid_3 = GridSearchCV(make_pipeline( preprocess_pipeline, RandomForestRegressor(n_jobs=-1, max_features= 'sqrt', n_estimators=50, oob_score = True)),
                    param_grid={
                   'randomforestregressor__n_estimators': [50, 150, 300],
                    'randomforestregressor__max_features': ['auto', 'sqrt', 'log2'],  
                    'randomforestregressor__max_depth': [ 100, 150, 200]
                   },
                    cv=kfold,
                    refit=True)
grid_3.fit(X_train, y_train)
print(grid_3.best_params_)
print(grid_3.best_estimator_)


```

```python
from sklearn import  metrics


models = []

models.append(('Lasso', grid_1.best_estimator_))
models.append(('Ridge', grid_2.best_estimator_))
models.append(('LR', grid_0.best_estimator_))
models.append(('RFR', grid_3.best_estimator_))


r2 = []
explained_variance_score = []
median_absolute_error = []
mean_squared_error = []
mean_absolute_error = []

for name, model in models:
    print(name)
    print("R^2: {}".format(metrics.r2_score(y_test, model.predict(X_test)) ))
    print("Explained variance score: {}".format( metrics.explained_variance_score(y_test, model.predict(X_test)) ))
    print("Median absolute error: {}".format( metrics.median_absolute_error(y_test, model.predict(X_test)) ))
    print("Mean squared error: {}".format( metrics.mean_squared_error(y_test, model.predict(X_test)) ))
    print("Mean absolute errors: {}".format(metrics.mean_absolute_error(y_test, model.predict(X_test)) ))
    
    r2.append(metrics.r2_score(y_test, model.predict(X_test)))
    
    explained_variance_score.append(metrics.explained_variance_score(y_test, model.predict(X_test)))
    median_absolute_error.append( metrics.median_absolute_error(y_test, model.predict(X_test)))
    mean_squared_error.append(metrics.mean_squared_error(y_test, model.predict(X_test)))
    mean_absolute_error.append(metrics.mean_absolute_error(y_test, model.predict(X_test)))
```

```python
d = {'r2': r2, 
     'explained_variance_score': explained_variance_score, 
     'median_absolute_error': median_absolute_error,
     'mean_squared_error' : mean_squared_error,
     'mean_absolute_error' : mean_absolute_error,
    }
df = pd.DataFrame(data=d)
df.insert(loc=0, column='Method', value=['Lasso','Ridge','LR','RFR'])
df
```
